```python
df["text"]=df["text"].apply(lambda x : re.sub("https?://[A-Za-z0-9.-/]+[A-Z0-9a-z]+", "",x))
df["text"].to_csv("text_file1.csv", index=False)
```

```python
def mention(x):
  fd=re.findall("@\w+", x)
  if  fd:
    return fd
  return None
```

```python
arr=df["text"].apply(lambda x : mention(x))
arr
```

```python
mentions=[]
for x in arr:
  if x!=None:
    mentions.extend(x)

mentions[: 10]

mentions_count=pd.Series(mentions).value_counts().head(100)
mentions_count
```
