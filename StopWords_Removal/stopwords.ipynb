# Stop Words

```
file=open("moon.txt", mode='r');
text=file.read();
file.close();
```

```
print(text)
```

Cleaning Text


```
import re
def clean_text(text):
    text=text.lower();
    text=re.sub('[,.\n]', '', text);
    text=re.sub("-", " ", text);
    return text


cleaned_text=clean_text(text);
```

```
print(cleaned_text)
```

finding most frequent words

```
import spacy
nlp=spacy.load("en_core_web_sm");
doc=nlp(cleaned_text);
```

```
#creating dictionary


words_dict={}
for token in doc:
  if token.text in words_dict:
    words_dict[token.text]=words_dict[token.text]+1;
  else:
    words_dict[token.text]=1;
```

```
import pandas as pd
df=pd.DataFrame({'words':list(words_dict.keys()), 'count':list(words_dict.values())});
df.sort_values(by='count', ascending=False, inplace=True);
print(df.shape)
```

```
df.head()
```

```
print(nlp.Defaults.stop_words)
```

```
new_tokens=[token.text for token in doc if(token.is_stop==False)]
```

```
print(new_tokens)
```

```
new_words={}
for token in new_tokens:
  if token in new_words:
    new_words[token]=new_words[token]+1;
  else:
    new_words[token]=1;

```

```
new_df=pd.DataFrame({"words": list(new_words.keys()), "count":list(new_words.values())})

new_df.sort_values(by="count", ascending=False, inplace=True)

print(new_df.shape)
new_df.head()
```

# Stop Words Exercise

```
file=open("switzerland.txt", mode='r')
text=file.read()
file.close()
print(text)
```

```
def clean_text(text):
  text=text.lower();
  text=re.sub('[,.\n]', '', text);
  text=re.sub("-", " ", text);
  return text
cleaned_text=clean_text(text);
print(cleaned_text)
```

```
import spacy
nlp1=spacy.load("en_core_web_sm");
doc1=nlp1(cleaned_text);
```

```
w_dict={}
for token in doc1:
  if token.text in w_dict:
    w_dict[token.text]=w_dict[token.text]+1;
  else:
    w_dict[token.text]=1;
```

```
import pandas as  pd
df1=pd.DataFrame({"words": list(w_dict.keys()), "count": list(w_dict.values())})
df1.sort_values(by="count", ascending=False, inplace=True)
print(df1.shape)
df1.head()
```

```
stop_words=[token.text for token in doc1 if(token.is_stop==True)]
print(stop_words)
```

```
length_stop=len(stop_words)
length_doc=len(doc1);
print(length_stop)
print(length_doc)
per=(length_stop/length_doc)*100;
print(per)
```

```
new_tokens1=[token.text for token in doc1 if(token.is_stop==False)]
print(new_tokens1)
print(len(new_tokens1))
```

```
w_dict1={}
for token in new_tokens1:
  if token in w_dict1:
    w_dict1[token]=w_dict1[token]+1;
  else:
    w_dict1[token]=1;

print(w_dict1)
```

```
dataf=pd.DataFrame({"words": list(w_dict1.keys()), "count": list(w_dict1.values())})
dataf.sort_values(by="count", ascending=False, inplace=True)
print(dataf.shape)
dataf.head()
```
